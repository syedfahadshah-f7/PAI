```python
#task 1
import matplotlib.pyplot as plt
student_heights = [152, 158, 165, 170, 175, 160, 162, 168, 155, 180,
                   177, 169, 173, 163, 172, 174, 164, 161, 167, 171]

students = [
    "Alice Johnson",
    "Bob Smith",
    "Charlie Brown",
    "Daisy Miller",
    "Ethan Davis",
    "Fiona White",
    "George Black",
    "Hannah Green",
    "Ian Thompson",
    "Julia Wilson",
    "Kevin Lee",
    "Liam Clark",
    "Mia Lewis",
    "Noah Walker",
    "Olivia Hall",
    "Peter Allen",
    "Quinn Young",
    "Rita Adams",
    "Sam Taylor",
    "Tina Harris"
]

colors = ['g','b','r','y','m','black','pink']

plt.figure(figsize=(12, 6))
plt.bar(students, student_heights, color = colors)
plt.xlabel("Students")
plt.ylabel("Height (cm)")
plt.title("Student Heights")
plt.xticks(rotation=90)
plt.show()
plt.pie(student_heights, labels =students,colors = colors, autopct= "%1.1f%%")
plt.show()
```

```python
#task 4
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("/content/heart .csv")

# Define the age grouping function
def apply(age):
    if 18 <= age <= 25:
        return "Young"
    elif 26 <= age <= 30:
        return "Mid Age"
    elif 31 <= age <= 35:
        return "Mature"
    else:
        return "Old"


df["Age_Group"] = df["age"].apply(apply)

df['Age_Group'].value_counts().plot(kind="pie", autopct='%1.1f%%')
plt.title("Age Group Distribution")
plt.show()

```

```python
#task 8
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]

y1 = [120, 150, 170, 160, 180, 200]

y2 = [100, 130, 160, 150, 190, 210]

plt.plot(x, y1, label='Line 1' ,color = "pink", marker = "o")
plt.plot(x, y2, label='Line 2', color = "grey",marker = "o")
plt.xlabel('Amazing X-axis')
plt.ylabel('Amazing Y-axis”')
plt.title("Two Lines on One Graph")
plt.legend()
plt.show()
```

```python
#task 7
import matplotlib.pyplot as plt

years = list(range(1920, 2021, 5))
sea_level_rise = [
    0, 1, 2, 2.5, 3, 3.2, 4, 4.5, 5, 5.8,
    6, 6.5, 7, 7.2, 7.5, 8, 8.2, 8.5, 9, 9.5,9.7
]

plt.figure(figsize=(10, 6))
plt.scatter(years, sea_level_rise, color='blue', s=100)


plt.title('Sea Level Rise Over the Past 100 Years')
plt.xlabel('Year')
plt.ylabel('Sea Level Rise (mm)')
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Iris.csv")
print(df.head())
```

```python
# Line Plot

x = df["SepalLengthCm"]
y = df["SepalWidthCm"]

plt.plot(x,y, linestyle = "--", marker = "o", mfc = "red")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
# scatter Plot

x = df["SepalLengthCm"]
y = df["SepalWidthCm"]

plt.scatter(x,y, marker = "o", edgecolors= "black", linewidths= 2, alpha = 0.6)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
#bar Plot

y = df["Species"].value_counts()
x = ["A","B","C"]
plt.bar(x, y, width=0.5, color = "r", edgecolor = "black", linewidth = 2)
plt.xlabel("Species")
plt.ylabel("Counts")
plt.show()
```

```python
# hist Plot

x = df["SepalLengthCm"]

plt.hist(x, bins=10,width = 0.3, color="r", edgecolor="black", linewidth =2.4)
plt.xlabel("Species")
plt.ylabel("Frequency")
plt.show()
```

```python
# Pie plot

y = df["Species"].value_counts()
x = ["A","B","C"]
plt.pie(y, labels = x, autopct ="%1.2f%%")
plt.legend()
plt.xlabel("Species")
plt.ylabel("Counts")
plt.show()
```

```python
# pair plot

sns.pairplot(df, hue="Species")
plt.show()
```

```python
# Heat Map
# df2 = df[["SepalLengthCm",  "SepalWidthCm" , "PetalLengthCm",  "PetalWidthCm"]]
import pandas as pd
import seaborn as sns
df = pd.read_csv("Iris.csv")
sns.heatmap(data = df.select_dtypes(include="number").corr(), annot=True)
plt.show()
```

```python
#Violin plot

sns.violinplot(x="Species", y="SepalLengthCm", data=df, color="y")
plt.show()
```

```python
#Box plot
import seaborn as sns
sns.boxplot(x= [450,123,567,444,350,780,909,678,567])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6])
y = np.array([1,2,3,4,5,6])

plt.plot(x,y,color="r")
plt.fill_between(x,y,color='g',where=(x>=2) & (x<=4))
plt.grid()
plt.show()
```
